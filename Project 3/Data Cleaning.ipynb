```python
# importing necessary libraries

import pandas as pd
import numpy as np
```

```python
#importing file

data=pd.read_csv("AB_NYC_2019.csv")
data
```

```python
#removing any duplicates 

data=data.drop_duplicates()
data
```

```python
#deleting columns that would be irrelevant for analytics

data=data.drop(columns=["name","neighbourhood","calculated_host_listings_count"])
data.head()
```

```python
# checking for null values ,dtype in all columns

data.info()
```

# Cleaning Host_name Column

```python
#converting dtype and removing nan values

data["host_name"]=data["host_name"].astype(str)
data["host_name"]=data["host_name"].str.replace("nan"," ")

#removing all non-alphabetical characters
data["host_name"]=data["host_name"].str.replace("[^a-zA-Z]","",regex=True)

#creating space between any 2 names

data["host_name"]=data["host_name"].str.replace(r"([a-z])([A-Z])",r"\1\2",regex=True)
data.head()
```

```python
# obtaining data where the AirBnb is priced

data_priced=data[data["price"]!=0]
data_priced.head()
```

```python
# if we need data for where minimum nights is a year or less

data_year=data[data["minimum_nights"]<=367]
data_year.head()
```

```python
# to get data where available is atleast 1 day

data_available=data[data["availability_365"]!=0]
data_available.head()
```

# Cleaning Last Review Column

```python
# converting data_type to string

data["last_review"]=data["last_review"].astype(str)

```

```python
# Deleting nan values from last review column

data["last_review"]=data["last_review"].str.replace("nan"," ")
data.head()
```

# Cleaning Reviews/Month Column

```python
#converting data_type to string

data["reviews_per_month"]=data["reviews_per_month"].astype(str)
```

```python
#deleting nan values from reviews per month column

data["reviews_per_month"]=data["reviews_per_month"].str.replace("nan"," ")
data.head()
```

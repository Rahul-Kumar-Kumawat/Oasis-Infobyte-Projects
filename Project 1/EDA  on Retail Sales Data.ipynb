```python
#Ordering Data by Date so to do Time Series Analytics

data=data.sort_values(by="Date")
data.head(5)
```

```python
#Importing Matplotlib And Seaborn

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Converting Date To Date-Time format

data["Date"]=pd.to_datetime(data["Date"])
```

```python
# Obtaining a plot of Total Amount Of Particular Dates with Seaborn

plt.figure(figsize=(15,5))
sns.lineplot(x="Date",y="Total Amount",data=data)
plt.show()
```

# Time Series Analysis By Month

```python
# Creating Month And Date Columns

data["Month"]=data["Date"].dt.month
data['Year']=data['Date'].dt.year
```

```python
# Calculating Total Monthly Sales

Monthly_sales=data.groupby(["Year","Month"],as_index=False)["Total Amount"].sum()
data["Monthly_sales"]=Monthly_sales["Total Amount"]
```

```python
# A plot of Total_monthly Sales Throughout the year

plt.figure(figsize=(8,5))
sns.lineplot(x="Date",y="Monthly_sales",data=data)
plt.show()
```

```python
# Deleting Month ,Year and Monthly_sales columns for customer and product Analysis

data=data.drop(["Month","Year","Monthly_sales"],axis=1)
data.head(10)
```

# Customer And Product Analysis

```python
age_bins=[10,20,30,40,50,60]
age_labels=["10-19","20-29","30-39","40-49","50+"]
data["Age group"]=pd.cut(data["Age"],bins=age_bins,labels=age_labels,right=True)
```

```python
# Calculating Total Amount Spent By Each Age Group

Total_amount_by_age=data.groupby("Age group")["Total Amount"].sum()
print(Total_amount_by_age)
```

```python
#Grouping by age group and product category and calculating the sum of spending

Grouped_data=data.groupby(["Age group","Product Category"])["Total Amount"].sum().reset_index()
```

```python
#Creating a bar plot to visualize spending patterns by age group and product category

plt.figure(figsize=(8,6))
sns.barplot(x="Age group",y="Total Amount",hue="Product Category",data=Grouped_data)
plt.title("Spending Patterns By Age Group And Product Category")
plt.xlabel("Age Group")
plt.ylabel("Total Amount")
plt.show()
```

# Gender Spending Analysis

```python
# Calculating total amount spent by each gender

Gender_Totalspend=data.groupby("Gender")["Total Amount"].sum().reset_index()
print(Gender_Totalspend)
```

```python
# Calculating the total spend plot for each gender

plt.figure(figsize=(5,5))
plt.bar(Gender_Totalspend["Gender"],Gender_Totalspend["Total Amount"],color=["tomato","green"])
plt.title("Total Amount Spent By Each Gender")
plt.xlabel("Gender")
plt.ylabel("Total Amount")
plt.show()
```

```python
# product perference by gender

Gender_preference=data.groupby(["Gender","Product Category"])["Total Amount"].sum().reset_index()
```

```python
# A plot of Products showing  Preference by Total Amount Spent on each by each gender

plt.figure(figsize=(8,8))
sns.barplot(x="Gender",y="Total Amount",hue="Product Category",data=Gender_preference)
plt.title("Gender Product Prefernce")
plt.xlabel("Gender")
plt.ylabel("Total Amount")
plt.show()
```

```python
# Total Amount spent by Average Male and Female

Avg_amount_by_Gender=data.groupby("Gender")['Total Amount'].mean()
print(Avg_amount_by_Gender)
```

# Product Performance Analysis

```python
# Total revenue from Each Product Category

Total_revenue=data.groupby("Product Category")["Total Amount"].sum().reset_index()
print(Total_revenue)
```

```python
plt.figure(figsize=(6,6))
plt.bar(Total_revenue["Product Category"],Total_revenue["Total Amount"],color=["g","r","b"])
plt.title("Total revenue from each Product")
plt.xlabel("Product Category")
plt.ylabel("Total Amount")
plt.show()
```

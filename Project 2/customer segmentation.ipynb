```python
# Assigning Education_status/level(1-5) to different rows

def Education_status(row):
    if row["education_Graduation"]==1:
        return "3"
    elif row["education_PhD"]==1:
        return "5"
    elif row["education_Master"]==1:
        return "4"
    elif row["education_Basic"]==1:
        return "1"
    elif row["education_2n Cycle"]==1:
        return "2"
    else:
        return "Unknown"
data["Education_status"]=data.apply(Education_status,axis=1)
data.head()
```

# Segmentation

```python
#importing k-means and other libraries

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.cluster import KMeans
```

```python
# Standardising the scale of the different columns to be used in segmentation

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
cols_to_cluster=["Income","MnTotal","In_relationship","Education_status"]
data_scaled=data.copy()
data_scaled[cols_to_cluster]=scaler.fit_transform(data[cols_to_cluster])
```

```python
#choosing a range for cluster values for elbow method

x=data_scaled[cols_to_cluster]
inertia_list=[]
for k in range(1,11):
    inertia=KMeans(n_clusters=k,n_init=10).fit(x).inertia_
    inertia_list.append(inertia)
```

```python
# visualization of the elbow method plot

plt.figure(figsize=[7,6])
plt.plot(range(1,11),inertia_list)
plt.title("Inertia vs Number of Clusters")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.show()

#4 appears to be the ideal of clusters
```

```python
# fitting desired number of clusters into our model and creating the segments

model =KMeans(n_clusters=4,random_state=7)
model.fit(data_scaled[cols_to_cluster])
data_scaled["Cluster"]=model.predict(data_scaled[cols_to_cluster])
```

```python
# generating 2 most relevent components /Properties of the clusters

from sklearn import decomposition
pca=decomposition.PCA(n_components=2)
pca_res=pca.fit_transform(data_scaled[cols_to_cluster])
data_scaled["pc1"]=pca_res[:,0]
data_scaled["pc2"]=pca_res[:,1]
```

```python
#Visualization of the Clusters

plt.figure(figsize=(8,6))
sns.scatterplot(x="pc1",y="pc2",data=data_scaled,hue="Cluster",palette="viridis")
plt.title("Cluster Visualization")
plt.xlabel("Principal Component 1 (pc1)")
plt.ylabel("Principal Component 2 (pc2)")
plt.legend(title="Clusters")
plt.show()
```

# Properties Of The Clusters

```python
#Number of clusters per cluster

customer_num=data_scaled.groupby("Cluster")["MnTotal"].count().reset_index()

#visualization

plt.figure(figsize=(8,8))
sns.barplot(x="Cluster",y="MnTotal",data=customer_num,palette="viridis")

plt.title("Cluster sizes")
plt.xlabel("Cluster")
plt.ylabel("Customer Num")
plt.show()
```

```python
# Forming Clusters without modifying the Original values of columns used to form clusters

scaler=StandardScaler()
cols_to_cluster=["Income","MnTotal","In_relationship","Education_status"]
data_clustered=data.copy()

model=KMeans(n_clusters=4,random_state=7)
model.fit(data_clustered[cols_to_cluster])
data_clustered["Cluster"]=model.predict(data_clustered[cols_to_cluster])
data_clustered.head()
```

```python
# Average Income Of Each Cluster

data_clustered["Avg_income"]=data_clustered.groupby("Cluster")["Income"].mean()
```

```python
# Bar plot to visualize

plt.figure(figsize=(8,8))
sns.barplot(x="Cluster",y="Avg_income",data=data_clustered,palette="viridis")
plt.title("Cluster Avg_income")
plt.xlabel("Cluster")
plt.ylabel("Avg income")
plt.show()
```
